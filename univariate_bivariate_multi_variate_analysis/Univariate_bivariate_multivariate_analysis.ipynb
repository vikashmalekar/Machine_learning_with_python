```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df =pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
df
```

### Univariate Analysis

```python
df_class0 = df.loc[df["target"]==0]
```

```python
df_class0
```

```python
df_class1 = df.loc[df['target']==1]
```

```python
df_class2 = df.loc[df['target']==2]
```

```python
plt.plot(df_class0['sepal length (cm)'],np.zeros_like(df_class0['sepal length (cm)']),'o')
plt.plot(df_class1['sepal length (cm)'],np.zeros_like(df_class1['sepal length (cm)']),'o')
plt.plot(df_class2['sepal length (cm)'],np.zeros_like(df_class2['sepal length (cm)']),'o')
plt.xlabel('sepal length')
plt.show()
```

### Bivariate Analysis

```python
sns.FacetGrid(df,hue = 'target',size = 5).map(plt.scatter,'sepal length (cm)','sepal width (cm)').add_legend();
```

### Multi variate Analysis

```python
sns.pairplot(df,hue = 'target',size = 5);
```
