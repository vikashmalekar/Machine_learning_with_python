```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(-5,5,100)
```

```python
y = x**2
```

```python
def f(z):
    return z**2
```

```python
plt.plot(x,y)
plt.xlabel('z')
plt.ylabel('f(z)')
z = -3
plt.scatter(z,f(z),c='red')
plt.show()
```

```python
def slope(x):
    return 2*x
```

```python
z_update_list = []
learning_rate = 0.1
for i in range(10):
    z = z-learning_rate*slope(z)
    z_update_list.append(z)
print(z_update_list)
    
```

```python
plt.plot(x,y)
plt.xlabel('z')
plt.ylabel('f(z)')
for i in z_update_list:
    plt.scatter(i,f(i),c='red')
```

Here we can observe that how the function values is minminzed from one iteration to other iteration
