```python
x=df.drop("Class",axis=1)
y=df['Class']
xdata=x.values
ydata = y.values
print(xdata)
print(ydata)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
from sklearn import svm
```

```python
svm_model = svm.SVC(kernel = "linear")
svm_model.fit(xtrain,ytrain)
```

```python
ypre = svm_model.predict(xtest)
no_of_errors = len([ypre!=ytest])
no_of_errors
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print("accuracy:",accuracy_score(ytest,ypre))
print("Classification_report is",classification_report(ypre,ytest))
```

```python
conf_mat = confusion_matrix(ytest,ypre)
sns.heatmap(conf_mat,xticklabels = ["normal","fraud"],yticklabels = ["normal","fraud"])
plt.title("confusion matrix")
plt.xlabel("Predicted class")
plt.ylabel("True class")
plt.show()
```
